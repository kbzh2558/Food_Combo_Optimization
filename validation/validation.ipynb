```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('data/survey_validation_indiv.xlsx')
```

```python
# drop
data = data.drop(columns=['ID', 'Started on', 'Last updated on', 'Status', 'Collector',
       'Language', 'IP address', 'Device', 'Consent:','Lunch Combo Selection','Enter your Prolific ID to initiate the payment.'])

data = data[data['Attention']=='Somewhat satisfied'].rename(columns={'hot dogImage 12':'R.GRILL HOT DOG'})
data.head()
```

```python
# fill nan and x
data.iloc[:, 2:23] = data.iloc[:, 2:23].map(lambda x: 1 if x == 'x' else 0 if pd.isna(x) else x)
data.iloc[:, 2:23]
```

```python
# to select list
item_cols = data.columns[2:23]
data['chosen_items'] = data[item_cols].apply(lambda row: item_cols[row == 1].tolist(), axis=1)
data = data.drop(columns=list(item_cols)+['Attention'])
```

```python
combo_dict = {
    'combo1_size2': ['MUFFIN MORCEAUX CHOCOLAT', 'R.GRILL HOT DOG'],
    'combo2_size2': ['COLLATION FRUITS ET FROMAGE', 'PANINI AU STEAK'],
    'combo3_size3': ['COLLATION FRUITS ET FROMAGE', 'PANINI AU STEAK','RED BULL 250ML'],
    'combo4_size4': ['BAGUETTINE AU JAMBON', "LAY'S CLASSIC 66g", 'PEPSI 591ML','ESKA 500ML'],
    'combo5_size4': ['MUFFIN MORCEAUX CHOCOLAT', 'R.GRILL HOT DOG', 'RED BULL SANS SUCRE 250ML','MOYEN CAFE REGULIER'],
    'combo6_size4': ['BAGUETTINE AU JAMBON', 'CHOCOLATINE', 'RED BULL 250ML','PEPSI 591ML','EAU SOURCE 500ML'],
}

# col counting matching items
for combo_name, combo_items in combo_dict.items():
    data[combo_name] = data['chosen_items'].apply(lambda items: sum(item in combo_items for item in items))
```

```python
# to % and bins
import re

for col in combo_dict.keys():
    # extract combo name and size
    match = re.match(r"(combo\d+)_size(\d+)", col)
    if match:
        combo_name = match.group(1)
        combo_size = int(match.group(2))
        
        # create % column
        pct_col = f"{combo_name}_percent"
        data[pct_col] = data[col] / combo_size * 100
        
        # bins
        bin_col = f"{combo_name}_engagement_level"
        data[bin_col] = pd.cut(
            data[pct_col],
            bins=[-1, 1, 33, 55, 100],
            labels=['None', 'Low', 'Medium', 'High']
        )
```

```python
import nltk
import spacy
from textblob import TextBlob
from nltk.corpus import stopwords
import spacy

# load
nlp = spacy.load("en_core_web_sm")

# def key words
vege_keywords = [
    'vegetarian', 'vegan', 'veggie', 'vege', 'plant-based',
    'plant based', 'meatless', 'no meat', 'vegatarian', 'veg', 'vgt', 'meat','processed meat', 'processed','animal products','trust'
]

# func process text
def clean_and_flag(text):
    if pd.isna(text):
        return "", [], 0

    doc = nlp(text)

    # cleaned tokens -> lowercase, alpha, no stop words
    tokens = [token.text.lower() for token in doc if token.is_alpha and not token.is_stop]

    # noun phrases
    phrases = [chunk.text.lower() for chunk in doc.noun_chunks]

    # vege mention flag
    vege_flag = int(any(re.search(rf"\b{kw}\b", text.lower()) for kw in vege_keywords))

    return ' '.join(tokens), phrases, vege_flag

other_cols = data.columns[data.columns.str.lower().str.startswith('other')]

for col in other_cols:
    result = data[col].apply(lambda x: pd.Series(clean_and_flag(x)))
    result.columns = [f"{col}_cleaned", f"{col}_phrases", f"{col}_is_vege"]
    data = pd.concat([data, result], axis=1)
```

```python
# unite to one flag
data[data.columns[data.columns.str.endswith('_is_vege')]].max(axis=1)

data['is_vege_any'] = data[data.columns[data.columns.str.endswith('_is_vege')]].max(axis=1)

data = data.drop(columns=data.columns[data.columns.str.endswith('_is_vege')])
```

```python
from collections import Counter

# find phrase cols
phrase_cols = data.columns[data.columns.str.endswith('_phrases')]

# flatten
all_phrases = []

for col in phrase_cols:
    col_phrases = data[col].dropna().tolist()
    for phrase_list in col_phrases:
        all_phrases.extend(phrase_list)

# count freq
phrase_counts = Counter(all_phrases)
phrase_df = pd.DataFrame(phrase_counts.items(), columns=['Phrase', 'Count'])
# sort
phrase_df = phrase_df.sort_values(by='Count', ascending=False).reset_index(drop=True)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering

# top only
top_phrase_df = phrase_df[phrase_df['Count'] >= 2].copy()  # or adjust threshold as needed
top_phrases = top_phrase_df['Phrase'].tolist()


# embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(top_phrases)

# cluster
clustering = AgglomerativeClustering(
    n_clusters=None, 
    distance_threshold=0.4, 
    linkage='ward'
)
labels = clustering.fit_predict(embeddings)
top_phrase_df['Cluster'] = labels

cluster_summary = top_phrase_df.groupby('Cluster').agg({
    'Phrase': list,
    'Count': sum
}).reset_index()


```

```python
cluster_summary
```

```python
# func nps and cate
def compute_nps_and_counts(series):
    total = series.notna().sum()
    promoters = (series >= 9).sum()
    passives = ((series >= 7) & (series <= 8)).sum()
    detractors = (series <= 6).sum()
    nps = ((promoters - detractors) / total) * 100 if total > 0 else None
    return promoters, passives, detractors, nps

# cols
rating_cols = list(data.iloc[:,1:7].columns)
combo_names = ['combo1', 'combo2', 'combo3', 'combo4', 'combo5', 'combo6']
spending_col = "How much do you anticipate to spend for lunch at a conveniece store like Couche Tard?"

# segment 
results = []

for i, rating_col in enumerate(rating_cols):
    combo = combo_names[i]
    engagement_col = f"{combo}_engagement_level"

    if engagement_col not in data.columns:
        continue  

    # groupings [spend lvl and engagement and is vege]
    grouped = data[[rating_col, spending_col, engagement_col,'is_vege_any']].dropna().groupby([spending_col, engagement_col,'is_vege_any'])

    for (spending_level, engagement_level, is_vege), group in grouped:
        promoters, passives, detractors, nps = compute_nps_and_counts(group[rating_col])
        results.append({
            'Rating Column': rating_col,
            'Spending Level': spending_level,
            'Engagement Level': engagement_level,
            'Is Vege': is_vege,
            'Promoters': promoters,
            'Passives': passives,
            'Detractors': detractors,
            'NPS': nps
        })

segmented_df = pd.DataFrame(results)

segmented_df
```

```python
# overall
overall_nps_df = (
    segmented_df
    .groupby('Rating Column')[['Promoters', 'Passives', 'Detractors']]
    .sum()
    .reset_index()
)

overall_nps_df['Total Responses'] = (
    overall_nps_df['Promoters'] +
    overall_nps_df['Passives'] +
    overall_nps_df['Detractors']
)

overall_nps_df['Overall NPS'] = (
    (overall_nps_df['Promoters'] - overall_nps_df['Detractors']) /
    overall_nps_df['Total Responses']
) * 100

overall_nps_df
```

```python
# segment by engagement lvl
engagement_lvl_nps_df = (
    segmented_df
    .groupby(['Rating Column','Engagement Level'])[['Promoters', 'Passives', 'Detractors']]
    .sum()
    .reset_index()
)

engagement_lvl_nps_df['Total Responses'] = (
    engagement_lvl_nps_df['Promoters'] +
    engagement_lvl_nps_df['Passives'] +
    engagement_lvl_nps_df['Detractors']
)

engagement_lvl_nps_df['Overall NPS'] = (
    (engagement_lvl_nps_df['Promoters'] - engagement_lvl_nps_df['Detractors']) /
    engagement_lvl_nps_df['Total Responses']
) * 100

engagement_lvl_nps_df
```

```python
# segment by Spending Level	
spending_lvl_nps_df = (
    segmented_df
    .groupby(['Rating Column','Spending Level'])[['Promoters', 'Passives', 'Detractors']]
    .sum()
    .reset_index()
)

spending_lvl_nps_df['Total Responses'] = (
    spending_lvl_nps_df['Promoters'] +
    spending_lvl_nps_df['Passives'] +
    spending_lvl_nps_df['Detractors']
)

spending_lvl_nps_df['Overall NPS'] = (
    (spending_lvl_nps_df['Promoters'] - spending_lvl_nps_df['Detractors']) /
    spending_lvl_nps_df['Total Responses']
) * 100

spending_lvl_nps_df
```

```python
# segment by engagement lvl and vege
vege_engagement_lvl_nps_df = (
    segmented_df
    .groupby(['Rating Column','Engagement Level', 'Is Vege'])[['Promoters', 'Passives', 'Detractors']]
    .sum()
    .reset_index()
)

vege_engagement_lvl_nps_df['Total Responses'] = (
    vege_engagement_lvl_nps_df['Promoters'] +
    vege_engagement_lvl_nps_df['Passives'] +
    vege_engagement_lvl_nps_df['Detractors']
)

vege_engagement_lvl_nps_df['Overall NPS'] = (
    (vege_engagement_lvl_nps_df['Promoters'] - vege_engagement_lvl_nps_df['Detractors']) /
    vege_engagement_lvl_nps_df['Total Responses']
) * 100

vege_engagement_lvl_nps_df
```

```python
# segment by spending and vege
vege_spending_lvl_nps_df = (
    segmented_df
    .groupby(['Rating Column','Spending Level', 'Is Vege'])[['Promoters', 'Passives', 'Detractors']]
    .sum()
    .reset_index()
)

vege_spending_lvl_nps_df['Total Responses'] = (
    vege_spending_lvl_nps_df['Promoters'] +
    vege_spending_lvl_nps_df['Passives'] +
    vege_spending_lvl_nps_df['Detractors']
)

vege_spending_lvl_nps_df['Overall NPS'] = (
    (vege_spending_lvl_nps_df['Promoters'] - vege_spending_lvl_nps_df['Detractors']) /
    vege_spending_lvl_nps_df['Total Responses']
) * 100

vege_spending_lvl_nps_df
```
